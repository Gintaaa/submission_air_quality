## Import Semua Packages/Library yang Digunakan

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Wrangling

### Gathering Data

Pada tahap ini, dilakukan penggabungan 12 dataset kualitas udara dari berbagai stasiun pemantau menjadi satu DataFrame tunggal untuk mempermudah proses analisis.

```python
# Tentukan path folder data
path = 'data'

file_list = [f for f in os.listdir(path) if f.endswith('.csv')]

print(f"Ditemukan {len(file_list)} file CSV:")
print(file_list)
```

```python
# List kosong untuk menampung dataframe dari setiap file
all_df = []

# Proses membaca dan memasukkan ke list
for file in file_list:
    file_path = os.path.join(path, file)
    temp_df = pd.read_csv(file_path)
    all_df.append(temp_df)

# Menggabungkan semua dataframe di dalam list menjadi satu
main_df = pd.concat(all_df, ignore_index=True)

# Menampilkan informasi ringkas tentang hasil penggabungan
print(f"Total baris data gabungan: {main_df.shape[0]}")
main_df.head()
```

**Insight:**
- Berhasil mengumpulkan dan menggabungkan total 420.768 baris data dari 12 file CSV yang berbeda (setiap file mewakili satu stasiun pemantau).
- Dataset memiliki informasi yang lengkap, mencakup parameter kualitas udara (PM2.5, PM10, SO2, NO2, CO, O3) dan kondisi meteorologi (TEMP, PRES, DEWP, RAIN, WSPM).
- Terlihat secara sekilas pada kolom polutan (seperti NO2 dan CO) terdapat nilai NaN (data kosong), yang menunjukkan perlunya proses Cleaning Data di tahap berikutnya.

### Assessing Data

```python
# Cek tipe data dan jumlah baris yang terisi
main_df.info()
```

**Insight:**

Kolom waktu masih terbagi-bagi menjadi 4 kolom terpisah dengan tipe data integer. Untuk mempermudah analisis tren waktu (time-series), kolom-kolom ini harus digabung menjadi satu kolom bertipe datetime.

```python
# cek jumlah total data yang kosong di setiap kolom
print(main_df.isna().sum())
```

**Insight:**

Terdapat missing values yang signifikan pada kolom polutan dan cuaca. Hal ini kemungkinan disebabkan oleh kerusakan sensor atau kegagalan sistem pada jam-jam tertentu. Data ini perlu ditangani agar tidak mengganggu perhitungan rata-rata.

```python
print("Jumlah duplikasi: ", main_df.duplicated().sum())
```

**Insight:**

Tidak ditemukan adanya data duplikat dalam dataset, sehingga setiap baris data mewakili catatan waktu yang unik.

```python
# Cek ringkasan parameter (rata-rata, nilai min, nilai max)
main_df.describe()
```

**Insight:**

Secara umum, distribusi data suhu dan cuaca berada pada rentang yang masuk akal. Nilai rata-rata PM2.5 yang tinggi menunjukkan adanya masalah polusi udara yang cukup serius di lokasi-lokasi tersebut.

### Cleaning Data

```python
# Menggunakan Forward Fill karena ini data deret waktu (time-series).
# Nilai yang kosong akan diisi dengan nilai dari jam sebelumnya yang tersedia.
# bfill setelah ffill untuk memastikan data di baris paling awal juga terisi dari jam setelahnya

main_df = main_df.ffill().bfill()
```

```python
# Gabungkan kolom year, month, day, hour menjadi satu kolom datetime tunggal.

main_df['datetime'] = pd.to_datetime(main_df[['year', 'month', 'day', 'hour']])
```

```python
# Kolom 'No' dan kolom waktu yang terpisah sudah tidak kita pakai lagi.

main_df.drop(columns=['No', 'year', 'month', 'day', 'hour'], inplace=True)
```

```python
# Pastikan lagi tidak ada data yang kosong dan tipe datanya sudah benar.

print("--- Hasil Akhir Cleaning ---")
main_df.info()
print("\nJumlah missing values sekarang:", main_df.isna().sum().sum())
```

**Insight:**
- Berhasil menangani seluruh data kosong menggunakan metode Forward Fill (ffill) dan Backward Fill (bfill).
- Berhasil menggabungkan kolom terpisah (year, month, day, hour) menjadi satu kolom tunggal datetime dengan tipe data yang benar (datetime64[ns]).
- Melakukan penghapusan kolom yang sudah tidak relevan (seperti kolom No dan kolom waktu yang lama) untuk mengoptimalkan memori dan fokus pada variabel analisis utama.

## Exploratory Data Analysis (EDA)

### Explore ...
Pada tahap ini, saya akan mengeksplorasi data untuk:

- Melihat rangkuman statistik nilai PM2.5 (rata-rata dan nilai ekstrem).

- Membandingkan kadar polusi antar stasiun untuk menemukan wilayah dengan kualitas udara terburuk.

- Melihat tren fluktuasi kadar polusi berdasarkan waktu (bulan dan musim).

```python
# Melihat rangkuman parameter statistik PM2.5 berdasarkan stasiun
main_df.groupby(by="station").agg({
    "PM2.5": ["mean", "max", "min", "std"]
}).sort_values(by=("PM2.5", "mean"), ascending=False)
```

```python
# Melihat rata-rata PM2.5 per bulan (1-12)
main_df.groupby(by=main_df.datetime.dt.month).agg({
    "PM2.5": "mean"
})
```

```python
# Membuat kolom season
def get_season(month):
    if month in [12, 1, 2]: return 'Winter'
    elif month in [3, 4, 5]: return 'Spring'
    elif month in [6, 7, 8]: return 'Summer'
    else: return 'Autumn'

main_df['season'] = main_df.datetime.dt.month.apply(get_season)

# Analisis rata-rata PM2.5 berdasarkan musim
main_df.groupby(by="season").agg({
    "PM2.5": "mean"
}).sort_values(by="PM2.5", ascending=False)
```

**Insight:**
- Stasiun Dongsi merupakan wilayah paling berpolusi dengan rata-rata PM2.5 sebesar 86.31.
- Stasiun Dingling memiliki kualitas udara terbaik dengan rata-rata terendah, yaitu 67.43.
- Ditemukan lonjakan polusi sangat parah di Wanshouxigong yang mencapai angka 999.0.
- Seluruh stasiun menunjukkan fluktuasi polusi yang tinggi (standar deviasi > 70), menandakan kualitas udara yang tidak stabil di semua lokasi.

## Visualization & Explanatory Analysis

### Pertanyaan 1:

Stasiun mana yang memiliki kualitas udara terburuk dan terbaik berdasarkan rata-rata kadar PM2.5 selama periode 2013-2017?

```python
import matplotlib.pyplot as plt
import seaborn as sns

station_rank = main_df.groupby("station")["PM2.5"].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(
    x="PM2.5", 
    y="station", 
    data=station_rank, 
    hue="station",
    legend=False,
    palette="Reds_r"
)
plt.title("Rata-rata Kadar PM2.5 per Stasiun (2013-2017)")
plt.xlabel("Kadar PM2.5 (µg/m³)")
plt.show()
```

**Insight:**
- Stasiun Terburuk: Stasiun Dongsi memiliki rata-rata konsentrasi PM2.5 paling tinggi, yang menunjukkan tingkat polusi udara paling parah di antara lokasi lainnya.
  
- Stasiun Terbaik: Stasiun Dingling secara konsisten menunjukkan angka PM2.5 terendah, menandakan wilayah ini memiliki kualitas udara yang paling bersih dibandingkan stasiun lainnya.

### Pertanyaan 2:

Bagaimana tren bulanan kadar PM2.5?

```python
# Grafik Tren Bulanan
monthly_avg = main_df.groupby(main_df.datetime.dt.month)["PM2.5"].mean()

plt.figure(figsize=(10, 5))
plt.plot(monthly_avg.index, monthly_avg.values, marker='o', linewidth=2, color="#72BCD4") 
plt.title("Tren Rata-rata PM2.5 per Bulan", loc="center", fontsize=15)
plt.xlabel("Bulan")
plt.ylabel("Rata-rata PM2.5")
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul', 'Agu', 'Sep', 'Okt', 'Nov', 'Des'])
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

**Insight:**
- Pola Fluktuasi: Kualitas udara tidak stabil sepanjang tahun; terdapat pola di mana kadar PM2.5 cenderung menurun secara signifikan pada pertengahan tahun, khususnya sekitar bulan Mei hingga Agustus.

- Kenaikan di Akhir Tahun: Konsentrasi polutan mulai merangkak naik tajam saat memasuki bulan November hingga mencapai puncaknya di sekitar bulan Januari atau Februari.

### Pertanyaan 3:

Bagaimana pola musiman polusi udara dan pada bulan apa polusi biasanya melonjak drastis?

```python
# Menyiapkan data
station_pm_df = main_df.groupby("station")["PM2.5"].mean().sort_values(ascending=False).reset_index()

# Menentukan warna: Merah tua untuk tertinggi, abu-abu untuk lainnya
colors = ["#D32F2F" if i == 0 else "#D3D3D3" for i in range(len(station_pm_df))]

plt.figure(figsize=(10, 5))
sns.barplot(x="PM2.5", y="station", data=station_pm_df, palette=colors, hue="station", legend=False)
plt.title("Stasiun Dongsi Mencatat Rata-rata Kadar PM2.5 Tertinggi", fontsize=15)
plt.xlabel("Konsentrasi PM2.5 (μg/m³)")
plt.ylabel("Stasiun")
plt.show()
```

**Insight:**
- Stasiun Dongsi memiliki polusi PM2.5 rata-rata tertinggi 86.3, sedangkan Dingling terendah 67.4.

- Ada selisih polusi yang cukup besar antar wilayah, menandakan adanya pengaruh aktivitas lokal yang berbeda di tiap stasiun.

```python
# Menggunakan 'ME' sesuai saran peringatan untuk standar terbaru
monthly_trend_df = main_df.resample(rule='ME', on='datetime')["PM2.5"].mean().reset_index()

plt.figure(figsize=(12, 5))
plt.plot(
    monthly_trend_df["datetime"], 
    monthly_trend_df["PM2.5"], 
    marker='o', 
    linewidth=2, 
    color="#3498db"
)

plt.title("Tren Bulanan Konsentrasi PM2.5 (2013-2017)", fontsize=15)
plt.xlabel("Tahun")
plt.ylabel("Rata-rata PM2.5 (μg/m³)")
plt.grid(True, alpha=0.3)
plt.show()
```

**Insight:**
- Polusi selalu melonjak tajam di awal dan akhir tahun (musim dingin) dan menurun di pertengahan tahun.
- Puncak polusi tertinggi terjadi pada awal 2017, dengan rata-rata bulanan melebihi 140.


## Analisis Lanjutan (Opsional)

ada bagian ini, saya melakukan pengelompokan kualitas udara berdasarkan konsentrasi PM2.5 menggunakan teknik Manual Grouping. Tujuannya adalah untuk mengklasifikasikan data ke dalam kategori tingkat polusi (Excellent hingga Heavily Polluted) guna memahami distribusi kualitas udara yang dihirup oleh masyarakat di area pemantauan.

```python
# Manual Grouping: Mengelompokkan kualitas udara berdasarkan PM2.5
def classify_air_quality(pm):
    if pm <= 35: return "Excellent"
    elif pm <= 75: return "Good"
    elif pm <= 115: return "Lightly Polluted"
    else: return "Heavily Polluted"

main_df['Quality_Category'] = main_df['PM2.5'].apply(classify_air_quality)

# Menampilkan distribusi kategori kualitas udara
print(main_df['Quality_Category'].value_counts())
```

- Sebagian besar kualitas udara tercatat pada kategori Good dan Lightly Polluted.

- Munculnya kategori Heavily Polluted menunjukkan adanya periode waktu tertentu di mana kualitas udara mencapai level yang berbahaya bagi kesehatan masyarakat.
